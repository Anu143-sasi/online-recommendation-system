```python

import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("my project.csv")

print(" Dataset loaded successfully.")
print(df.head())
# Drop duplicates if any
df.drop_duplicates(inplace=True)

# Handle missing values
df.fillna({
    'rating': df['rating'].mean(),
    'course_duration_hours': df['course_duration_hours'].median(),
    'feedback_score': df['feedback_score'].mean(),
    'difficulty_level': 'Intermediate',
    'certification_offered': 'No',
    'study_material_available': 'No'
}, inplace=True)
# Normalize column names (optional)
df.columns = [col.strip().lower().replace(' ', '_') for col in df.columns]

print("\n Cleaned dataset columns:")
print(df.columns)
```

Feature Transformation

```python
# Encode categorical features
df['certification_offered'] = df['certification_offered'].map({'Yes': 1, 'No': 0})
df['study_material_available'] = df['study_material_available'].map({'Yes': 1, 'No': 0})

# One-hot encode difficulty level
df = pd.get_dummies(df, columns=['difficulty_level'], drop_first=True)
```

 Feature Scaling (for numeric columns)


```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

num_cols = [
    'course_duration_hours', 'rating', 'enrollment_numbers',
    'course_price', 'feedback_score', 'time_spent_hours',
    'previous_courses_taken'
]
df[num_cols] = scaler.fit_transform(df[num_cols])

print("\n Scaled numerical features:")
print(df[num_cols].head())
```

 Create a Combined 'Course Profile' Feature (for Content-Based)


```python
df['course_profile'] = (
    df['course_name'].astype(str) + " " +
    df['instructor'].astype(str) + " " +
    df['difficulty_level_Intermediate'].astype(str) + " " +
    df['certification_offered'].astype(str) + " " +
    df['study_material_available'].astype(str)
)

print("\n Created 'course_profile' feature for content-based model.")
```

 Create User Behavior Features (for Collaborative Filtering)


```python
# Average rating per user
user_avg_rating = df.groupby('user_id')['rating'].mean().rename('user_avg_rating')

# Total courses taken per user
user_course_count = df.groupby('user_id')['course_id'].count().rename('user_course_count')

# Merge user-level stats
df = df.merge(user_avg_rating, on='user_id')
df = df.merge(user_course_count, on='user_id')
```

Create Course Popularity Features


```python
# Average rating per course
course_avg_rating = df.groupby('course_id')['rating'].mean().rename('course_avg_rating')

# Total enrollments per course
course_enrollments = df.groupby('course_id')['enrollment_numbers'].mean().rename('course_enrollments')

# Merge course-level stats
df = df.merge(course_avg_rating, on='course_id')
df = df.merge(course_enrollments, on='course_id')
```

 Final Feature-Engineered Dataset


```python
print("\n Final feature-engineered dataset preview:")
print(df.head())

# Save processed dataset
df.to_csv("feature_engineered_courses.csv", index=False)
print("\n Saved feature-engineered dataset as 'feature_engineered_courses.csv'")
```
